```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stats_07 = pd.read_csv('/Users/kabirdoshi/College/Research/image_skipping_exp/image_skipping/stats_0.7conf.csv')
stats_08 = pd.read_csv('/Users/kabirdoshi/College/Research/image_skipping_exp/image_skipping/stats_0.8conf.csv')
stats_09 = pd.read_csv('/Users/kabirdoshi/College/Research/image_skipping_exp/image_skipping/stats_0.9conf.csv')
stats_10 = pd.read_csv('/Users/kabirdoshi/College/Research/image_skipping_exp/image_skipping/stats_1.0conf.csv')

stats_07
```

```python
x=190
train_loss07 = stats_07.loc[0, :].values.flatten().tolist()[1:x]
train_loss08 = stats_08.loc[0, :].values.flatten().tolist()[1:x]
train_loss09 = stats_09.loc[0, :].values.flatten().tolist()[1:x]
train_loss10 = stats_10.loc[0, :].values.flatten().tolist()[1:x]

test_loss07 = stats_07.loc[1, :].values.flatten().tolist()[1:x]
test_loss08 = stats_08.loc[1, :].values.flatten().tolist()[1:x]
test_loss09 = stats_09.loc[1, :].values.flatten().tolist()[1:x]
test_loss10 = stats_10.loc[1, :].values.flatten().tolist()[1:x]

train_acc07 = stats_07.loc[2, :].values.flatten().tolist()[1:x]
train_acc08 = stats_08.loc[2, :].values.flatten().tolist()[1:x]
train_acc09 = stats_09.loc[2, :].values.flatten().tolist()[1:x]
train_acc10 = stats_10.loc[2, :].values.flatten().tolist()[1:x]

test_acc07 = stats_07.loc[3, :].values.flatten().tolist()[1:x]
test_acc08 = stats_08.loc[3, :].values.flatten().tolist()[1:x]
test_acc09 = stats_09.loc[3, :].values.flatten().tolist()[1:x]
test_acc10 = stats_10.loc[3, :].values.flatten().tolist()[1:x]





train_time07 = stats_07.loc[4, :].values.flatten().tolist()[1:x]
train_time08 = stats_08.loc[4, :].values.flatten().tolist()[1:x]
train_time09 = stats_09.loc[4, :].values.flatten().tolist()[1:x]
train_time10 = stats_10.loc[4, :].values.flatten().tolist()[1:x]

skip_rate07 = stats_07.loc[5, :].values.flatten().tolist()[1:x]
skip_rate08 = stats_08.loc[5, :].values.flatten().tolist()[1:x]
skip_rate09 = stats_09.loc[5, :].values.flatten().tolist()[1:x]
skip_rate10 = stats_10.loc[5, :].values.flatten().tolist()[1:x]

skip_rate07 = sum(skip_rate07) / (50000 * x)
skip_rate08 = sum(skip_rate08) / (50000 * x)
skip_rate09 = sum(skip_rate09) / (50000 * x)
skip_rate10 = sum(skip_rate10) / (50000 * x)


```

```python
plt.plot(train_loss07, label='train_loss_0.7')
plt.plot(train_loss08, label='train_loss_0.8')
plt.plot(train_loss09, label='train_loss_0.9')
plt.plot(train_loss10, label='train_loss_1.0')
plt.xlabel('epoch')
plt.ylabel('train_loss')
plt.title('train loss vs epoch')
plt.legend()
plt.show()
```

```python
plt.plot(test_loss07, label='test_loss_0.7')
plt.plot(test_loss08, label='test_loss_0.8')
plt.plot(test_loss09, label='test_loss_0.9')
plt.plot(test_loss10, label='test_loss_1.0')
plt.xlabel('epoch')
plt.ylabel('test_loss')
plt.title('test loss vs epoch')
plt.legend()
plt.savefig('test_loss_vs_epoch.png')
plt.show()

```

```python

plt.plot(train_acc07, label='train_acc_0.7')
plt.plot(train_acc08, label='train_acc_0.8')
plt.plot(train_acc09, label='train_acc_0.9')
plt.plot(train_acc10, label='train_acc_1.0')
plt.legend()
plt.show()
```

```python
plt.plot(test_acc07, label='test_acc_0.7')
plt.plot(test_acc08, label='test_acc_0.8')
plt.plot(test_acc09, label='test_acc_0.9')
plt.plot(test_acc10, label='test_acc_1.0')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('test_acc')
plt.title('Test Accuracy vs Epoch')
plt.savefig('testacc_vs_epoch.png')
plt.show()

```
